```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import MeanShift
MS = MeanShift(bandwidth=2).fit(x)
y_MS=MS.fit_predict(x)
```

```python
y_MS
```

```python
supervised=dataset
```

```python
supervised
```

```python
supervised["cluster_group"]=y_MS
```

```python
supervised
```

```python
supervised.to_csv("cluster.csv",index=False)
```

```python
import seaborn as sns
facet=sns.lmplot(data=supervised,
x=supervised.colums[3],
y=supervised.colums[4],
hue=supervised.colums[5],
fit_reg=True,
legend=True,
legend_out=True                 
```
